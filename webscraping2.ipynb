```python
# create list with all the rows
# remove thead class
# loop thru list get each row
# loop thru each row
# grab each rows data append those to a main list of that particular datatype
# its gonna be one big loop with sub-loops appending to main lists
```

```python
# left lists
coach = []
bowl = []
notes=[]

# center lists
preseasonRnk =[]
highestRnk = []
finalRnk = []

# right lists
season = []
wins= []
losses = []
winlosspct =[]
ptdiffsosched = []
sosched = []

mainsoup = soup2.select('tbody')[0]
# thead = mainsoup.select('.thead')
row = mainsoup.select('tr')
rowlist = []
rownot = []
```

```python
row[20].select('th')[0].getText()
```

```python

rowlist.clear()

# left lists
coach.clear()
bowl.clear()
notes.clear

# center lists
preseasonRnk =[]
highestRnk.clear()
finalRnk.clear()

# right lists
season.clear()
wins.clear
losses.clear()
winlosspct.clear
ptdiffsosched.clear()
sosched.clear()
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

# left lists
coach = []
bowl = []
notes=[]

# center lists
preseasonRnk =[]
highestRnk = []
finalRnk = []

# right lists
season = []
wins= []
losses = []
winlosspct =[]
ptdiffsosched = []
sosched = []

teamStat = [coach,bowl,notes,preseasonRnk,highestRnk,finalRnk,season,wins,losses,winlosspct,ptdiffsosched,sosched]

for x in range (len(row)):
    if row[x].select('th')[0].getText() == 'Rk':
        rownot.append(row[x])
    else:
        rowlist.append(row[x])
        
for xx in range(len(rowlist)):
    leftsub = rowlist[xx].select('.left') 
    
    coach.append(leftsub[1].getText())
    bowl.append(leftsub[2].getText())
    notes.append(leftsub[3].getText())

    centersub = rowlist[xx].select('.center')
    
    preseasonRnk.append(centersub[0].getText())
    highestRnk.append(centersub[1].getText())
    finalRnk.append(centersub[2].getText())
    
    rightsub = rowlist[xx].select('.right')

    season.append(rightsub[1].getText()) 
    wins.append(rightsub[2].getText())
    losses.append(rightsub[3].getText())
    winlosspct.append(rightsub[4].getText())
    ptdiffsosched.append(rightsub[5].getText())
    sosched.append(rightsub[6].getText())
    
```

```python
x= 2
for d in range (len(teamStat)):
    print(teamStat[d][x])
```
