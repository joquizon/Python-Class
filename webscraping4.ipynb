```python
import requests
import bs4
```

```python
baseUrl = 'http://www.feedbooks.com/books/top?category=FBFIC028000&lang=en&page={}'
```

```python
getter = (baseUrl.format(1)) 
result = requests.get(getter)
multisoup = bs4.BeautifulSoup(result.text, 'lxml')
bookcontainx = multisoup.select(".span-19")[1]
unitcount = len(bookcontainx.select('.space'))
print(bookcontainx.select('div .span-3')[4].select('img')[0]['src'])
print(bookcontainx.select('div .span-11')[2].select('h3')[0].text)
print(bookcontainx.select('div .span-11')[2].select('h4 .gray')[0].text)
print(bookcontainx.select('div .span-3')[5].select('h4 a')[0]['href'])
print(bookcontainx.select('div .span-3')[5].select('.small p')[0].text)
```

```python
class Bookscraped:
    def __init__(self,bookcover,booktitle,bookauthor,dlink,categ):
        self.bookcover = bookcover
        self.booktitle = booktitle
        self.bookauthor = bookauthor
        self.dlink = dlink 
        self.categ = categ
booked.clear()
booked=[]
book = {}
for z in range (56):
    getter = (baseUrl.format(z)) 
    result = requests.get(getter)
    multisoup = bs4.BeautifulSoup(result.text, 'lxml')
    bookcontainx = multisoup.select(".span-19")[1]
    unitcount = len(bookcontainx.select('.space'))

    for x in range(unitcount):
    # grabs book cover img
        bookcover = bookcontainx.select('div .span-3')[x*2].select('img')[0]['src']
    # grabs title
        booktitle = bookcontainx.select('div .span-11')[x].select('h3')[0].text
    # grabs author    
        bookauthor = bookcontainx.select('div .span-11')[x].select('h4 .gray')[0].text
    # grab dl link
        dlink = bookcontainx.select('div .span-3')[(x*2)+1].select('h4 a')[0]['href']
    # grab categ
        cats = bookcontainx.select('div .span-3')[(x*2)+1].select('.small p')
    # grabs specific categ from scrape instantiates it creates dict
        for kitty in range(len(cats)):
            if cats[kitty].text == 'Short Stories':
                categ = cats[kitty].text
                booked.append(Bookscraped(bookcover,booktitle,bookauthor,dlink,categ).__dict__)
            else:
                pass
    print('endloop')
print('big endloop')
print(booked)
```
